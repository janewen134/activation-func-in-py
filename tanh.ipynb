activation function -- tanh

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# prepare the data
input = np.linspace(-100,100,10)
# tanh = sinh(x)/ cosh(x) = e^x − e^(−x)/ e^x + e^(−x)
output = np.tanh(input)
```

```python
# plot
plt.plot(input,output)
plt.xlabel("tanh Inputs")
plt.ylabel("tanh Outputs")
plt.show()
```
