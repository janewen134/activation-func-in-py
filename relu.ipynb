activation function -- relu

```python
# define function
def relu(X):
   return np.maximum(0,X)
```

```python
# prepare data
import numpy as np
import matplotlib.pyplot as plt
input = np.linspace(-100,100)
output = relu(input)
```

```python
# plot
plt.plot(input, output)
plt.xlabel("relu inputs")
plt.ylabel("relu inputs")
plt.savefig('relu.png')
plt.show()
```
